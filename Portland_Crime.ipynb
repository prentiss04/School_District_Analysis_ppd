```python
# Read the crime data file and store it in a Pandas DataFrame. 
crime_data_df = pd.read_csv(PDX_crime_data_to_load)
crime_data_df.head()
```

```python
# Get a count of rows with the DF in order ot determine if there are any null values
crime_data_df.count()
```

```python
# Determine if there are any missing values in the crime data. 
crime_data_df.isnull().sum()
```

```python
# Drop null rows 
crime_data_df = crime_data_df[pd.isnotnull(crime_data_df['Address'])]
```

```python
no_null_crime_df = crime_data_df.dropna(how='any')
no_null_crime_df.head()    
```

```python
# Verify counts
no_null_crime_df.count()
```

```python
# Search through the "Offense Type" column and replace any similar values with one consistent value
set(no_null_crime_df['Offense Type'].tolist())
```

```python
# Combining similar offenses
no_null_crime_df = no_null_crime_df.replace({"Commercial Sex Acts": "Sex Stuff, 
                                             "Assisting or Promoting Prostitution": "Sex Stuff"})
no_null_crime_df
```
